```python
import pandas as pd
import numpy as np
filename='medicare_edit.csv'
data = pd.read_csv(filename,encoding= 'utf-8')
data
```

```python
data_medicare = [[0 for i in range(len(data.columns))] for j in range(3)]
data_medicare
    
```

```python
data_medicare[1][0]= '/'
data_medicare[2][0]= 'Nan'

for i in range(1,len(data.columns)):
    data_medicare[0][i]=data.columns[i]
    slash_count=0
    nan_count=0
    total=0
    for item in data[data.columns[i]]:
        if item == '/' or item == '\\':
            slash_count+=1
        elif isinstance(item,float) and np.isnan(item):
            nan_count+=1
        total+=1
    data_medicare[1][i]=slash_count/total
    data_medicare[2][i]=nan_count/total

data_medicare
```

```python
data_medicare=pd.DataFrame(data_medicare[1:],columns=data_medicare[0])
data_medicare
```

```python
data_medicare.T.to_csv('data_medicare.csv',encoding='utf-8')
```
