```python
data_d=pd.read_csv('cut/cut_disease/cut_text_list.csv')
data_d=data_d['投保范围_投保人符合条件集合']
text=''
for i in data_d:
    text+=i

text=text.replace('[','').replace(']',',').replace('\'','').replace(',',' ').replace('、',' ')
```

```python
import jieba
seg_list=jieba.cut(text)
res=[]
for i in list(seg_list):
    if i!=' ' and i!='或' and i!='不详':
        res.append(i)
res=pd.Series(res)
res_stats=res.value_counts()
```

```python
wordcloud=pyecharts.WordCloud('疾病_投保范围_投保人符合条件集合',width=1000, height=1000)
wordcloud.add('',res_stats.index,res_stats,word_size_range=[20,150])
```

```python
data_d=pd.read_csv('cut/cut_medicare/cut_text_list.csv')
data_d=data_d['投保范围_投保人符合条件集合']
text=''
for i in data_d:
    text+=i

text=text.replace('[','').replace(']',',').replace('\'','').replace(',',' ').replace('、',' ')
seg_list=jieba.cut(text)
res=[]
for i in list(seg_list):
    if i!=' ' and  i!='的' and i!='或' and i!=',':
        res.append(i)
res=pd.Series(res)
res_stats=res.value_counts()

wordcloud=pyecharts.WordCloud('医疗_投保范围_投保人符合条件集合',width=2000, height=1500)
wordcloud.add('',res_stats.index,res_stats,word_size_range=[20,150])
```
